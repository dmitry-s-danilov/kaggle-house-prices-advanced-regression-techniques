### Variables Descriptions

```python
from pandas import set_option, reset_option
from contest.data.variables.description import construct

set_option(
    'display.max_rows', None,
    'display.max_columns', None,
)
```

#### Construct Description

```python
description = construct()

description
```

#### Categorical Variables

```python
type_ = 'categorical'
column = 'type'

description_ = description[
    description[
        column
    ].apply(
        # lambda _: _[0] == type_
        lambda _: type_ in _
    )
].sort_values(
    by=column
)

description_
```

##### Nominal Variables

```python
# type_ = ('categorical', 'nominal')
type_ = 'nominal'
column = 'type'
columns = [
    'description',
    'categories',
]

description_ = description.loc[
    # description[column] == type_,
    description[
        column
    ].apply(
        # lambda _: _[1] == type_
        lambda _: type_ in _
    ),
    columns
]

description_
```

##### Ordinal Variables
